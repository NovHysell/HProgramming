```python
from pathlib import Path
import json
a = input("Input a directory: ")
b = input("Input another directory: ")
a = Path(a)
b = Path(b)
list_a = []
list_b = []
list_c = []
try:
    for file in a.iterdir():
        if file.is_file():
            list_a.append(file.name)
except:
    print(f"Error: Directory not found: {a}")
try:
    for file in b.iterdir():
        if file.is_file():
            list_b.append(file.name)
except:
    print(f"Error: Directory not found: {b}")
for file in list_b:
    for test in list_a:
        if file == test:
            list_a.remove(file)
            list_b.remove(file)
            list_c.append(file)
list_a.sort()
list_b.sort()
list_c.sort()
print
step1 = "DIRECTORY DIFF REPORT"
step2 = "===================="
step3 = f"DIR A: /path/to/{a.name}\nDIR B: /path/to/{b.name}"
step4 = f"FILES ONLY IN A:\n"
for test in list_a:
    step4 = step4 + f"- {test}\n"
step5 = f"FILES ONLY IN B:\n"
for test in list_b:
    step5 = step5 + f"- {test}\n"
step6 = f"COMMON FILES:\n"
for test in list_c:
    step6 = step6 + f"- {test}\n"
result = f"{step1}\n{step2}\n\n{step3}\n\n{step4}\n{step5}\n{step6}"
with open("diff_report.txt", "w") as file:
    file.write(result)
```

```python
import json
from fractions import Fraction

class FractionEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, Fraction):
            return {
                "__fraction__": True,
                "numerator": obj.numerator,
                "denominator": obj.denominator
            }

data = {
    "questions": [
        {"id": 1, "value": Fraction(3, 4)},
        {"id": 2, "value": Fraction(2, 6)},
        {"id": 3, "value": Fraction(-5, -10)}
    ],
    "answer": Fraction(5, 8)
}

with open("fractions.json", "w") as f:
    json.dump(data, f, indent=4, cls=FractionEncoder)

print(f"Success! Data is saved in {f.name}")
```

```python
import json
from fractions import Fraction

def fraction_decoder(obj):
    if isinstance(obj, dict):
        cl = obj.get("__fraction__", None)
        if cl == True:
            num = obj["numerator"]
            den = obj["denominator"]
            return Fraction(num, den)
    return obj

# === TEST CODE ===
with open("fractions.json", "r") as f:
    loaded_data = json.load(f, object_hook=fraction_decoder)

data = loaded_data.get("questions", 0)
for d in data:
    print(d.get("value", 0))
```
